## 2.1. Contexto de la base de datos

Este dataset proviene de un estudio **clínico** realizado en la década de **1990**, diseñado para investigar y comprender mejor la progresión del **SIDA** en pacientes que recibieron diferentes tipos de tratamientos **antirretrovirales**. Los datos fueron recolectados para analizar cómo diversos factores personales, históricos y médicos influyen en la probabilidad de infección y en la **progresión** de la enfermedad.

La base de datos incluye información detallada sobre el tiempo hasta un evento crítico o la censura de los pacientes, así como múltiples variables relacionadas con la demografía, el historial médico, el tipo de tratamiento recibido y los resultados de laboratorio. Entre estas variables se encuentran la edad, el peso, la raza, el género, la actividad sexual, el historial de uso de drogas intravenosas, el estado funcional medido por la **escala de Karnofsky**, y varios indicadores de tratamientos y síntomas.

El objetivo principal de este conjunto de datos es proporcionar a los investigadores un recurso para estudiar la efectividad de diferentes regímenes de tratamiento, entender los factores de riesgo asociados con la infección por el virus del SIDA y desarrollar modelos predictivos que puedan ayudar a personalizar el tratamiento y mejorar los resultados en pacientes con VIH/SIDA.

# Diccionario de los datos

## La insuficiencia hepática (secundario)


*   es causada por un trastorno o una sustancia que daña el hígado.

*   En la mayoría de los casos se presenta ictericia (coloración amarillenta en la piel y los ojos), se sienten cansancio y debilidad y se pierde el apetito.

*   Otros síntomas incluyen la acumulación de líquido en el abdomen (ascitis) y la tendencia a la formación de hematomas y a sangrar con facilidad.

*   El médico generalmente diagnostica la insuficiencia hepática basándose en los síntomas, en la exploración y en los análisis de sangre.

*   El tratamiento consiste generalmente en controlar el consumo de proteínas, restringir el sodio en la dieta, evitar el alcohol por completo y tratar la causa, pero a veces es necesario un trasplante hepático.

## **`trt` - Indicador de Tratamiento**


   - **¿Qué es?**
     Esta variable indica el tipo de tratamiento antirretroviral que recibió un paciente al inicio del estudio. Los tratamientos están codificados con los valores 0, 1, 2, y 3, que corresponden a combinaciones específicas de medicamentos.

   - **Valores:**
     - **0: ZDV solo (Zidovudina)**  
       La Zidovudina (ZDV) es uno de los primeros medicamentos antirretrovirales utilizados para tratar el VIH. Actúa inhibiendo la replicación del virus, lo que ayuda a reducir la carga viral en el cuerpo del paciente. ZDV solo es una opción de tratamiento básica y puede ser utilizada en etapas iniciales o en combinación con otros medicamentos.
     - **1: ZDV + ddI (Zidovudina + Didanosina)**  
       Esta combinación incluye Zidovudina y Didanosina (ddI). La Didanosina también inhibe la replicación del VIH, y su uso junto con ZDV puede ofrecer un efecto sinérgico, mejorando la eficacia del tratamiento y ayudando a retrasar la progresión de la enfermedad.
     - **2: ZDV + Zal (Zidovudina + Zalcitabina)**  
       Este régimen combina Zidovudina con Zalcitabina (Zal), otro inhibidor de la transcriptasa inversa. La combinación se utiliza para atacar el virus desde diferentes frentes, lo que puede ser más efectivo en ciertos pacientes o en etapas más avanzadas de la enfermedad.
     - **3: ddI solo (Didanosina)**  
       Didanosina, utilizada sola, es otra opción de tratamiento antirretroviral que inhibe la replicación del VIH. Este tratamiento puede ser elegido por razones de tolerancia, resistencia a otros medicamentos, o preferencia clínica basada en el perfil del paciente.

## **`cd40` - Conteo de CD4 al Inicio del Estudio**


   - **¿Qué es?:**
     El conteo de células CD4 al inicio del estudio es una medida clave del estado del sistema inmunológico de un paciente. Las células CD4 son un tipo de glóbulo blanco que juega un papel fundamental en la defensa del cuerpo contra infecciones.
   - **Valores:**
        Números enteros que representan la cantidad de células CD4 por microlitro de sangre.
   - **Contexto:**
     Un conteo alto de CD4 sugiere un sistema inmunológico fuerte, mientras que un conteo bajo indica una mayor vulnerabilidad a infecciones y enfermedades oportunistas, características del VIH/SIDA. El conteo de CD4 es una de las métricas más utilizadas para monitorear la salud de un paciente con VIH y para determinar cuándo iniciar o cambiar el tratamiento antirretroviral.

### **`cd420` - Conteo de CD4 a las 20±5 Semanas**
   - **¿Qué es?:**
     Similar al `cd40`, esta variable mide el conteo de células CD4, pero en un punto específico del tiempo: aproximadamente 20 semanas después del inicio del estudio.
   - **Valores:**
        Números enteros que representan la cantidad de células CD4 por microlitro de sangre.
   - **Contexto:**
     El seguimiento del conteo de CD4 después de varias semanas de tratamiento permite evaluar la eficacia del tratamiento antirretroviral. Un aumento en el conteo de CD4 sugiere que el tratamiento está funcionando, mejorando la capacidad del cuerpo para combatir infecciones. Por otro lado, un conteo estancado o en descenso puede indicar la necesidad de ajustar el tratamiento.

##  **`cd80` - Conteo de CD8 al Inicio del Estudio**


   - **¿Qué es?:**
        También llamadas células T citotóxicas, son otro tipo de glóbulo blanco que puede destruir células infectadas por virus y células cancerosas.
   - **Valores:**
        Números enteros que representan la cantidad de células CD8 por microlitro de sangre.
   - **Contexto:**
     Un alto conteo de CD8 puede indicar que el sistema inmunológico está activo y combatiendo infecciones, mientras que un bajo conteo puede reflejar un sistema inmunológico comprometido. La relación entre los conteos de CD4 y CD8 es a menudo analizada para comprender mejor la respuesta inmunitaria de un paciente con VIH.

###  **`cd820` - Conteo de CD8 a las 20±5 Semanas**
   - **¿Qué es?:**
     Esta variable mide el conteo de células CD8 aproximadamente 20 semanas después del inicio del estudio.
   - **Valores:**
        Números enteros que representan la cantidad de células CD8 por microlitro de sangre.
   - **Contexto:**
     Similar al seguimiento de CD4, el conteo de CD8 a las 20 semanas es útil para evaluar la respuesta inmunológica a lo largo del tiempo y determinar la efectividad del tratamiento. Un cambio en el conteo de CD8 puede proporcionar información adicional sobre cómo el cuerpo del paciente está manejando el virus y cómo está respondiendo al tratamiento.


# VIH y SIDA

### Resaltar
*   El VIH ataca y destruye las células CD4 (linfocitos CD4) del sistema inmunitario que combaten las infecciones. La pérdida de linfocitos CD4 dificulta la lucha del cuerpo contra las infecciones y ciertas clases de cáncer. Sin tratamiento, el VIH puede gradualmente destruir el sistema inmunitario, lo que causa deterioro de la salud y la aparición del SIDA. Con tratamiento, el sistema inmunitario se puede recuperar.


### Recuento
*   Un recuento normal de células **CD4** es de alrededor de 600 a 1.500 células.
*   Un recuento normal de células **CD8** es de alrededor de 300 a 1.000 células.


### Importante:

*   El valor de **CD8** es menor a 150

### Prueba de hipotesis  Para la variable de Karnofsky

#### Hipótesis Nula (H0):

No Hay una diferencia en la puntuación media de Karnofsky entre los pacientes infectados y no infectados por SIDA.
$$\begin{aligned}
H_0&: \mu\ infectados = \mu\ No infectados\\
\end{aligned}$$

#### Hipótesis alternativa (H₁):

Hay una diferencia en la puntuación media de Karnofsky entre los pacientes infectados y no infectados por SIDA.
$$\begin{aligned}
H_1&: \mu\ infectados ≠ \mu\ No infectados  .
\end{aligned}$$

````python
from scipy.stats import norm
norm.ppf(0.05)
````

````python
# Calcular la media y la varianza
mean_infectados = np.mean(infectadosk)
mean_no_infectados = np.mean(noinfectadosk)
print(f"Media infectados: {mean_infectados}")
print(f"Media no infectados: {mean_no_infectados}")
var_infectados = np.var(infectadosk, ddof=1)
var_no_infectados = np.var(noinfectadosk, ddof=1)
print(f"Varianza infectados: {var_infectados}")
print(f"Varianza no infectados: {var_no_infectados}")
# Tamaños de muestra
n_infectados = len(infectadosk)
n_no_infectados = len(noinfectadosk)
print(f"Tamaño de muestra infectados: {n_infectados}")
print(f"Tamaño de muestra no infectados: {n_no_infectados}")

sdv_infectados =  infectadosk.std()
sdv_no_infectados = noinfectadosk.std()
print(f"Desviación estandar infectados: {sdv_infectados}")
print(f"Desviación estandar no infectados: {sdv_no_infectados}")
````

````python
# Error
error_estandar = math.sqrt((var_infectados / n_infectados) + (var_no_infectados / n_no_infectados))

# Estadístico de prueba z
z = (mean_infectados - mean_no_infectados) / error_estandar
print(f"Estadístico de prueba: z = {z}")

# Cálculo del p-valor
#1.1641161812292466e-08= 0.0000000116
p_valor = 2 * (1 - norm.cdf(abs(z)))  # Bilateral
print(f"Valor-p: {p_valor}")

# Comparación con alfa
alpha = 0.05

if p_valor < alpha:
    print("Rechazamos H0")
else:
    print("No rechazamos H0")

# Valor crítico
z_alpha = norm.ppf(1 - alpha / 2)  # Para prueba bilateral
print(f"Valor crítico: z_alpha = {z_alpha}")
````

### Entropía de variables

````python
# Definir la variable objetivo
target_column = 'infected'

# Calcular la entropía
def calcular_entropia(series):
    probs = series.value_counts(normalize=True)
    return entropy(probs, base=2)

# Calcular la entropía inicial de la variable objetivo
entropia_inicial = calcular_entropia(df_5k[target_column])
print(f"Entropía inicial de '{target_column}': {entropia_inicial:.4f}")

# Variables predictoras
variables = [col for col in df_5k.columns if col != target_column]

# Evaluar cada variable
mejor_variable = None
max_ganancia = -float('inf')

for variable in variables:
    entropia_condicional = 0
    for valor in df_5k[variable].unique():
        subset = df_5k[df_5k[variable] == valor]
        prob = len(subset) / len(df_5k)
        entropia_condicional += prob * calcular_entropia(subset[target_column])

    ganancia_informacion = entropia_inicial - entropia_condicional
    print(f"Variable: {variable} | Entropía condicional: {entropia_condicional:.4f} | Ganancia de información: {ganancia_informacion:.4f}")

    if ganancia_informacion > max_ganancia:
        max_ganancia = ganancia_informacion
        mejor_variable = variable

# Resultado final
print(f"\nLa mejor variable para dividir es: {mejor_variable} con una ganancia de información de {max_ganancia:.4f}")
````

````python
# Definir la variable objetivo
target_column = 'infected'

# Calcular la entropía
def calcular_entropia(series):
    probs = series.value_counts(normalize=True)
    return entropy(probs, base=2)

# Calcular la entropía inicial de la variable objetivo
entropia_inicial = calcular_entropia(df_5k[target_column])
print(f"Entropía inicial de '{target_column}': {entropia_inicial:.4f}")

# Variables predictoras
variables = [col for col in df_5k.columns if col != target_column]

# Evaluar cada variable y almacenar los resultados
resultados = []

for variable in variables:
    entropia_condicional = 0
    for valor in df_5k[variable].unique():
        subset = df_5k[df_5k[variable] == valor]
        prob = len(subset) / len(df_5k)
        entropia_condicional += prob * calcular_entropia(subset[target_column])

    ganancia_informacion = entropia_inicial - entropia_condicional
    resultados.append({"Variable": variable,
                       "Entropía Condicional": entropia_condicional,
                       "Ganancia de Información": ganancia_informacion})

# Crear un dataFrame con los resultados y ordenarlo
df_resultados = pd.DataFrame(resultados)
df_resultados = df_resultados.sort_values(by="Ganancia de Información", ascending=False)

# Mostrar las mejores variables en orden
print("\nRanking de las mejores variables para dividir:")
print(df_resultados)
````

### Entrenamiento

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, f1_score

# Variables ordenadas según el ranking de entropía
selected_features = ['wtkg', 'cd80', 'preanti', 'cd820', 'time', 'cd420', 'cd40', 'strat','str2']

X = df_5k[selected_features]
#X = df_5k.drop(columns=['infected'])
y = df_5k['infected']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el árbol
modelo = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=42,class_weight='balanced')
modelo.fit(X_train, y_train)
````

````python
# Predicción en el conjunto de prueba
y_pred = modelo.predict(X_test)

# Evaluación del modelo
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1-Score:", f1_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
````

### Matriz de confusion

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Realizar predicciones en todo el conjunto de datos
y_pred_total = modelo.predict(X)  # X contiene todas las características del dataset

# Obtener la matriz de confusión
cm = confusion_matrix(y, y_pred_total)

# Extraer los valores de la matriz de confusión
TP = cm[1, 1]  # Verdaderos Positivos
TN = cm[0, 0]  # Verdaderos Negativos
FP = cm[0, 1]  # Falsos Positivos
FN = cm[1, 0]  # Falsos Negativos

# Total de ejemplos
total = cm.sum()

# Calcular los porcentajes
TP_percentage = TP / total * 100
TN_percentage = TN / total * 100
FP_percentage = FP / total * 100
FN_percentage = FN / total * 100

# Mostrar los resultados
print(f"Verdaderos Positivos (TP): {TP} ({TP_percentage:.2f}%)")
print(f"Verdaderos Negativos (TN): {TN} ({TN_percentage:.2f}%)")
print(f"Falsos Positivos (FP): {FP} ({FP_percentage:.2f}%)")
print(f"Falsos Negativos (FN): {FN} ({FN_percentage:.2f}%)")

# Crear las etiquetas que combinan los valores y los porcentajes
cm_display = cm.astype('str')  # Convertir los valores a cadena
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        cm_display[i, j] = f"{cm[i, j]} ({(cm[i, j] / total * 100):.2f}%)"

# Graficar la matriz de confusión
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=cm_display, fmt='s', cmap='Purples', xticklabels=['No Infectado', 'Infectado'], yticklabels=['No Infectado', 'Infectado'])
plt.title("Matriz de Confusión con Porcentajes y Números")
plt.ylabel("Real")
plt.xlabel("Predicción")
plt.show()
````

### Porcentaje aciertos y errores

````python
# Realizar predicciones en todo el conjunto de datos
y_pred_total = modelo.predict(X)  # X contiene todas las características del dataset
aciertos_totales = sum(y_pred_total == y)  # Comparar con los valores reales
total_ejemplos = len(y)  # Total de ejemplos en el dataset

# Calcular porcentaje de aciertos y errores
porcentaje_aciertos = (aciertos_totales / total_ejemplos) * 100
porcentaje_errores = 100 - porcentaje_aciertos

# Datos para la gráfica
categorias = ['Aciertos (%)', 'Errores (%)']
valores = [porcentaje_aciertos, porcentaje_errores]

# Crear la gráfica de barras
import matplotlib.pyplot as plt

plt.bar(categorias, valores, color=['purple', 'pink'])
plt.title('Porcentaje de Aciertos y Errores en la Muestra Completa')
plt.ylabel('Porcentaje (%)')
plt.ylim(0, 100)  # Escala para porcentajes
plt.text(0, porcentaje_aciertos + 1, f"{porcentaje_aciertos:.2f}%", ha='center', fontsize=12)
plt.text(1, porcentaje_errores + 1, f"{porcentaje_errores:.2f}%", ha='center', fontsize=12)

# Mostrar la gráfica
plt.show()

# Imprimir resultados en texto
print(f"Total de ejemplos: {total_ejemplos}")
print(f"Aciertos totales: {aciertos_totales}")
print(f"Porcentaje de aciertos: {porcentaje_aciertos:.2f}%")
print(f"Porcentaje de errores: {porcentaje_errores:.2f}%")
````
